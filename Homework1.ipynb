Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
a = float(input("Nhap chieu rong:"))
b = float(input("Nhap chieu dai:"))
if a > b:
    print("Gia tri nhap khong phu hop")
elif a<=0 or b<=0:
    print("Gia tri nhap khong phu hop")
else:
    dt_hv = a * b
    dt_ht = 3.14 * (a/2)**2
    dt_tc = dt_hv - dt_ht
    print(f"Dien tich trong cay la {dt_tc:.2f}")
```

Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma)

```python
c = input()
if c.isupper() == True:
    print(f"{c.lower()}")
else:
    print(f"{c.upper()}")
```

Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt.

```python
c = input()
if c == "A":
    print(f"{c} la truong hop dac biet")
else:
    ma_hoa = ord(c)
    thuong = chr(ma_hoa + 31)
    print(f"{thuong}")
```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”

```python
a,b,c = map(float,input().split())
if a+b>c and a+c>b and b+c>a:
    p = (a+b+c)/2
    S = (p*(p-a)*(p-b)*(p-c)) ** 0.5
    print(f"Dien tich tam giac la {S:.1f}")
else:
    print("Khong 3 canh cua tam giac")
```
