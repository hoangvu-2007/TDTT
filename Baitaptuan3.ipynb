## W3A3

Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

## W3A11 [Phân loại tam giác]

lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường".

```python
a,b,c = map(int,input().split())
if a>0 and b>0 and c>0:
    if a+b > c and a+c>b and b+c>a:
        if a == b == c:
            print("Tam giác đều")
        elif a == b or a == c or b == c:
            print("Tam giác cân")
        else:
            print("Tam giác thường")
    else:
        print("Không phải tam giác")
else:
    print("Không phải tam giác")
```

- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh

```python
n = int(input())
if 0<=n<=50:
    tien = n*1500
elif 51<=n<=100:
    tien = 50*1500 + 2000*(n-50)
else:
    tien = 50*1500 + 50*2000 + 3000*(n-100)
print(tien)
```

## W3A16 [Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm
sẵn có của pyton). Các kết quả cách nhau một dấu cách

```python
n = float(input())
phan_nguyen = int(n)
phan_thap_phan = n - phan_nguyen
if n >= 0:
    if phan_thap_phan == 0:
        tron_len = phan_nguyen
        tron_xuong = phan_nguyen
    else:
        tron_len = phan_nguyen + 1
        tron_xuong = phan_nguyen   
    if phan_thap_phan >= 0.5:
        tron_gan_nhat = phan_nguyen + 1
    else:
        tron_gan_nhat = phan_nguyen
else:
    if phan_thap_phan == 0:
        tron_len = phan_nguyen
        tron_xuong = phan_nguyen
    else:
        tron_len = phan_nguyen
        tron_xuong = phan_nguyen - 1    
    if phan_thap_phan <= -0.5:
        tron_gan_nhat = phan_nguyen - 1
    else:
        tron_gan_nhat = phan_nguyen
print(f"{tron_len} {tron_xuong} {tron_gan_nhat}")
```

## W3A17 Cấp số nhân
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng
thứ tự a, b, c, d hay không?
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs
d

```python
a, b, c, d = map(int, input().split())
la_cap_so_nhan = False
if a == 0:
    if b == 0 and c == 0 and d == 0:
        la_cap_so_nhan = True
else:
    if b % a == 0:
        q = b // a 
        if (b * q == c) and (c * q == d):
            la_cap_so_nhan = True
if la_cap_so_nhan:
    print("YES")
else:
    print("NO")
```
